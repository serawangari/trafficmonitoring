```
from ultralytics import YOLO

DATA_YAML = '/content/drive/MyDrive/traffic_project/dataset_yolo/data.yaml'
MODEL = 'yolov8n.pt'  # change if needed

model = YOLO(MODEL)
model.train(data=DATA_YAML, epochs=50, imgsz=640)
```

```
model.val(data=DATA_YAML)
```
